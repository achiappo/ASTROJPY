```python
samples = np.load('../results/mcmc/cov/samples_cov10_n1.npy')
lnprobs = np.load('../results/mcmc/cov/lnprobs_cov10_n1.npy')
```

```python
Jmc = samples[:,0]
Lmc = -lnprobs
Lmin = min(Lmc)

Jenv, Lenv = [], []
hist, binsedges = np.histogram(Jmc, bins=30)
for Jl,Jr in zip( binsedges[:-1], binsedges[1:] ):
    indbin = np.where( np.logical_and( Jl<Jmc, Jmc<Jr ) )[0]
    if any(indbin):
        Lbin, Jbin = Lmc[indbin], Jmc[indbin]
        indLmin = np.where( Lbin==min(Lbin) )[0][0]
        Jenv.append( Jbin[indLmin] )
        Lenv.append( Lbin[indLmin] )

interpLenv = interp(Jenv, Lenv)
eval_Like_J = np.linspace(min(Jenv), max(Jenv), 100)
min_Like_J = min(interpLenv(eval_Like_J))
Jmin = eval_Like_J[ np.where( interpLenv(eval_Like_J) == min_Like_J )[0][0] ]
Jmin
```

```python
import numpy as np
import pylab as plt
from corner import hist2d

%matplotlib inline
```

```python
results = np.load('../results/mcmc/samples_Lr_J1672_s100.npy')
#lnprobs = np.load('../results/mcmc/lnprobs_Lrc_J1672_N10000.npy')
parameters = ['r', 'c'] #['r', 'a', 'b', 'c']

samples = results[:,:-1]
lnprobs = results[:, -1]
```

```python
truevals = [-0.46, 1]#, 3, 1]
fig, axes = plt.subplots(nrows=5, ncols=3, figsize=(16,26))
for n,N in enumerate( [1, 5, 10, 20, 50]):
    samples = np.array( [results[i*N,:-1] for i in range(len(results)/N)] )
    lnprobs = np.array( [results[i*N,-1] for i in range(len(results)/N)] )
    
    Lmin = min(-lnprobs)
    indLmin = np.where( -lnprobs==Lmin )[0][0]
    ind2Dhist = np.where( -lnprobs<Lmin+0.2 )[0]
    for p,param in enumerate( parameters ):
        values = samples[:,p]
        exec( "{} = values".format(param) )
        exec( "{}_min = values[indLmin]".format(param) )
        
    for a,ax in enumerate( axes.flat[n*3:(n+1)*3] ):
        if a==0:
            hist2d(x=samples[ind2Dhist, 0], 
                   y=samples[ind2Dhist, 1], 
                   ax=ax)
            ax.set_xlabel(parameters[0], fontsize=12);
            ax.set_ylabel(parameters[1], fontsize=12);
            ax.vlines(x=eval( '{}_min'.format(parameters[0]) ), 
                      ymin=eval( 'min({})'.format(parameters[1]) ),
                      ymax=eval( 'max({})'.format(parameters[1]) ),
                      color='m');
            ax.hlines(y=eval( '{}_min'.format(parameters[1]) ), 
                      xmin=eval( 'min({})'.format(parameters[0]) ),
                      xmax=eval( 'max({})'.format(parameters[0]) ),
                      color='c');
            ax.set_title('thin = %i'%N, fontsize=12);
        else:
            ax.plot(eval('{}[ind2Dhist]'.format(parameters[a-1])), 
                    -lnprobs[ind2Dhist], 
                    alpha=0.2, 
                    'bo');
            exec( 'parmin = {}_min'.format(parameters[a-1]) )
            ax.plot(parmin, 
                    Lmin, 
                    '%sd'%('m' if a==1 else 'c'), 
                    label= '%s = %.2f'%(parameters[a-1], parmin), 
                    ms=10 );
            ax.vlines(x=truevals[a-1], 
                      ymin=0, 
                      ymax=2000, 
                      label='true param: %.2f'%truevals[a-1]);
            ax.set_xlabel(parameters[a-1]);
            ax.set_ylabel('logLike', fontsize=12)
            ax.set_ylim(Lmin-0.1, Lmin+0.3);
            ax.legend(numpoints=1, loc='upper right');
plt.tight_layout()
#plt.savefig('../results/mcmc/thin_full_s100.png', dpi=300);
```
